```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
```

```python
df= pd.read_csv("Salary_Data.csv")
```

```python
plt.figure(figsize=(20 , 10));

plt.plot(df.YearsExperience , df.Salary , color= "#5F157E")
plt.title("Salary for every Experience Level" , color= "#5F157E" , fontsize= 20 , fontstyle= "italic")

plt.xlabel("Salary" , fontsize= 15 , fontstyle= "italic", color= "#5F157E")
plt.ylabel("Level" , fontsize= 15 , fontstyle= "italic", color= "#5F157E")

plt.show()
```

```python
plt.figure(figsize=(20 , 10));

plt.scatter(df.YearsExperience , df.Salary , color= "#5F157E")
plt.title("Relation Between Salary and Experience Level" , color= "#5F157E" , fontsize= 20 , fontstyle= "italic")

plt.xlabel("Salary" , fontsize= 15 , fontstyle= "italic", color= "#5F157E")
plt.ylabel("Level" , fontsize= 15 , fontstyle= "italic", color= "#5F157E")

plt.show()
```

```python
Y= df.iloc[: , -1]

X= df.iloc[: , :-1]
```

```python
print(X , Y)
```

```python
# salary(y)= model.coef_(m) *YearExperience (x) + model.intercept_ (b)
```

```python
model= LinearRegression()
```

```python
model.fit(X, Y)
```

```python
model.coef_  #------> M   
```

```python
model.intercept_ #-------> b
```

```python
model.predict([[6]])
```

```python
plt.figure(figsize=(14 , 8))
plt.scatter(X, Y)
plt.plot(X , model.predict(X), color= "#FFE400")
```

```python
model.score(X , Y)
```
